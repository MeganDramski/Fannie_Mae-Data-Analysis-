```python
# load Accusition data 

dfAcq = pd.read_table('/Users/meganpolak/Downloads/2016Q1/Acquisition_2016Q1.txt', names=( 'LOAN_ID','ORIG_CHN', 'Seller_Name', 'ORIG_RT','ORIG_AMT','ORIG_TRM','ORIG_DTE','FRST_DTE','OLTV','OCLTV','NUM_BO','DTI','CSCORE_B','FTHB_FLG','PURPOSE','PROP_TYP','NUM_UNIT','OCC_STATE','ZIP_3','MI_PCT','PRODUCT_TYPE','SCORE_C','MI_TYPE','RELOCATION_FLAG'),sep="|", index_col=False,infer_datetime_format=True )
```

```python
# drop all unnecessary columns 
df=dfAcq.drop(['RELOCATION_FLAG','PRODUCT_TYPE','MI_TYPE','MI_PCT','OCLTV','SCORE_C','FTHB_FLG', 'PURPOSE','ORIG_DTE','FRST_DTE','ORIG_CHN','Seller_Name'], axis=1)
# set Loan id as index 
df.set_index('LOAN_ID')
```

```python
# fill na with mean in df 
df=df.fillna(df.mean())
```
